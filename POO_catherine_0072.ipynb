# lenguaje Python, crear la clase "animal" con los atributos "nombre", "raza" crear el compocitor animal, y las funciones ladrar(), correr(); crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos.

```python
# Definición de la clase
class Animal:
    # El constructor (compositor)
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función ladrar
    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau, guau!😒")

    # Función correr
    def correr(self):
        print(f"{self.nombre} está corriendo velozmente.😎")

    def morder(self):
        print(f"{self.nombre} esta mordiendo a la vecina.🤦‍♂️")

# --- Creación de los objetos (Instanciación) ---

# Creamos el objeto pitbull
pitbull = Animal("Thor", "Pitbull")

# Creamos el objeto chihuahua
chihuahua = Animal("Chispa", "Chihuahua")

# creamos el objeto perro callejero con datos
perrocallejero = Animal("solovino", "chafa")

# --- Uso de los objetos ---

print(f"Perro 1: {pitbull.nombre}, Raza: {pitbull.raza}")
pitbull.ladrar()
pitbull.correr()

print("-" * 20)

print(f"Perro 2: {perrocallejero.nombre}, Raza: {perrocallejero.raza}")
perrocallejero.ladrar()
perrocallejero.correr()
perrocallejero.morder()

print("-" * 20)

print(f"Perro 3: {chihuahua.nombre}, Raza: {chihuahua.raza}")
chihuahua.ladrar()
chihuahua.correr()
print("programa realizado por catherin rascon NL = 26")
```
